# Python List Methods

## Methods to Add Elements to a List
The following methods are used for adding new item(s) into a list:

| Sr.No. | Method | Description |
|--------|--------|-------------|
| 1 | `list.append(obj)` | Appends object `obj` to the list. |
| 2 | `list.extend(seq)` | Appends the contents of `seq` to the list. |
| 3 | `list.insert(index, obj)` | Inserts object `obj` into the list at the specified `index`. |



## Methods to Remove Elements from a List
The following methods are used for removing items from a list:

| Sr.No. | Method | Description |
|--------|--------|-------------|
| 1 | `list.clear()` | Clears all the contents of the list. |
| 2 | `list.pop(obj=list[-1])` | Removes and returns the last object or the object at the specified index from the list. |
| 3 | `list.remove(obj)` | Removes the first occurrence of object `obj` from the list. |


## Methods to Access Elements in a List
These methods are used for finding or counting items in a list:

| Sr.No. | Method | Description |
|--------|--------|-------------|
| 1 | `list.index(obj)` | Returns the lowest index in the list where `obj` appears. |
| 2 | `list.count(obj)` | Returns the count of how many times `obj` occurs in the list. |

---

## Copying and Ordering Methods
These methods are used for creating copies and arranging items in a list:

| Sr.No. | Method | Description |
|--------|--------|-------------|
| 1 | `list.copy()` | Returns a copy of the list object. |
| 2 | `list.sort([func])` | Sorts the objects in the list in place, using a comparison function if provided. |
| 3 | `list.reverse()` | Reverses the order of objects in the list in place. |
