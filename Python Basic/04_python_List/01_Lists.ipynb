### Python Lists

List is one of the built-in data types in Python. A Python list is a sequence of comma separated items, enclosed in square brackets [ ]. The items in a Python list need not be of the same data type

List is an ordered collection of items. Each item in a list has a unique position index, starting from 0.

A list in Python is similar to an array in C, C++ or Java. However, the major difference is that in C/C++/Java, the array elements must be of same type. On the other hand, Python lists may have objects of different data types.


A Python list is mutable. Any item from the list can be accessed using its index, and can be modified. One or more objects from the list can be removed or added. A list may have same item at more than one index positions.

```python
# Accessing Values in Lists

list1 = ['physics', 'chemistry', 1997, 2000];
list2 = [1, 2, 3, 4, 5, 6, 7 ];
print ("list1[0]: ", list1[0])
print ("list2[1:5]: ", list2[1:5])
```

```python
# Updating Lists

list = ['physics', 'chemistry', 1997, 2000];
print ("Value available at index 2 : ")
print (list[2])
list[2] = 2001;
print ("New value available at index 2 : ")
print (list[2])
```

```python
# Delete List Elements 

list1 = ['physics', 'chemistry', 1997, 2000];
print (list1)
del list1[2];
print ("After deleting value at index 2 : ")
print (list1)
```

# Python List Operations

| **Python Expression**   | **Results**                 | **Description**        |
|--------------------------|-----------------------------|-------------------------|
| `[1, 2, 3] + [4, 5, 6]` | `[1, 2, 3, 4, 5, 6]`       | Concatenation           |
| `['Hi!'] * 4`           | `['Hi!', 'Hi!', 'Hi!', 'Hi!']` | Repetition             |
| `3 in [1, 2, 3]`        | `True`                      | Membership check        |


# Python List: Indexing and Slicing Table

| Python Expression | Result                    | Description                               |
|------------------|--------------------------|-------------------------------------------|
| `L[0]`           | `'spam'`                 | First element (index 0)                   |
| `L[1]`           | `'Spam'`                 | Second element                            |
| `L[2]`           | `'SPAM!'`                | Third element                             |
| `L[-1]`          | `'SPAM!'`                | Last element (negative index)             |
| `L[-2]`          | `'Spam'`                 | Second last element                        |
| `L[-3]`          | `'spam'`                 | Third last element                         |
| `L[0:2]`         | `['spam', 'Spam']`       | Slice from index 0 to 1 (end-1)          |
| `L[1:3]`         | `['Spam', 'SPAM!']`      | Slice from index 1 to 2                    |
| `L[:2]`          | `['spam', 'Spam']`       | Slice from start to index 1               |
| `L[1:]`          | `['Spam', 'SPAM!']`      | Slice from index 1 to end                 |
| `L[:]`           | `['spam', 'Spam', 'SPAM!']` | Full copy of the list                  |
| `L[::2]`         | `['spam', 'SPAM!']`      | Every second element (step=2)             |
| `L[::1]`         | `['spam', 'Spam', 'SPAM!']` | Step of 1 (normal copy)                |
| `L[::-1]`        | `['SPAM!', 'Spam', 'spam']` | Reverse list                            |
| `L[1::-1]`       | `['Spam', 'spam']`       | Reverse from index 1 to 0                 |
| `L[2:0:-1]`      | `['SPAM!', 'Spam']`      | Reverse from index 2 down to 1            |
| `L[-1::-1]`      | `['SPAM!', 'Spam', 'spam']` | Reverse using negative start           |
| `L[-2::-1]`      | `['Spam', 'spam']`       | Reverse from second last element to start |
| `L[:-1]`         | `['spam', 'Spam']`       | Slice from start to one before last       |
