### Array 

An array is a container which can hold a fix number of items and these items should be of the same type. Each item stored in an array is called an element and they can be of any type including integers, floats, strings, etc.

These elements are stored at contiguous memory location. Each location of an element in an array has a numerical index starting from 0. These indices are used to identify and access the elements.

```python
# Creating Arrayu in Python 

import array as array_name 

obj = array_name.array(typecode[,initializer])
```

- typecode âˆ’ The typecode character used to speccify the type of elements in the array.

- initializer âˆ’ It is an optional value from which array is initialized. It must be a list, a bytes-like object, or iterable elements of the appropriate type.

# Accing array items



```python
# using indexing 

import array as arr 

numericArray = arr.array('i',[111,211,311,411,511])
print(numericArray[0])
print(numericArray[1])
print(numericArray[2])

# using interation 
print("============================")

for item in numericArray:
    print(item)

# using enumerate() function
print("============================")
for loc, val in enumerate(numericArray):
    print(f"index: {loc}, value: {val}")
```

### Python Add Array Items

```python
# using append() methods

import array as arr 

a = arr.array('i',[1,2,3])
a.append(2)
print(a)

# using insert methods 

a.insert(1,10)
print(a)

# using extend methods 

a.extend([1,2,3])
print(a)
```

# Removing the Array 


```python
import array as arr 

a= arr.array('i',[1,2,3,4,5])
print(a)

# removing arary usng remove method

a.remove(1)
print(a)
a.pop()  # defualt last index 
print(a)
```

# Adding and Removing Elements 

| Sr.No. | Method             | Description                                                                 |
|--------|--------------------|-----------------------------------------------------------------------------|
| 1      | `append(x)`        | Appends a new item with value `x` to the end of the array.                  |
| 2      | `extend(iterable)` | Appends items from `iterable` to the end of the array.                      |
| 3      | `insert(i, x)`     | Inserts a new item with value `x` before position `i`.                      |
| 4      | `pop([i])`         | Removes and returns the item at index `i`. If `i` is not specified, removes and returns the last item. |
| 5      | `remove(x)`        | Removes the first occurrence of `x` from the array.                         |

# information and utility Methods

| Sr.No. | Method                        | Description                                                                 |
|--------|-------------------------------|-----------------------------------------------------------------------------|
| 1      | `buffer_info()`               | Returns a tuple `(address, length)` giving the memory address and length of the buffer. |
| 2      | `count(x)`                    | Returns the number of occurrences of `x` in the array.                      |
| 3      | `index(x[, start[, stop]])`   | Returns the smallest index where `x` is found. Optional `start` and `stop` define a sub-range. |

# Mainpulating Array Elements 

| Sr.No. | Method        | Description                                                                 |
|--------|---------------|-----------------------------------------------------------------------------|
| 1      | `reverse()`   | Reverses the order of the items in the array.                               |
| 2      | `byteswap()`  | Byteswaps all items, useful for reading data from machines with different byte order. |

# Conversion Methods

| Sr.No. | Method               | Description                                                                 |
|--------|----------------------|-----------------------------------------------------------------------------|
| 1      | `frombytes(buffer)`  | Appends items from a bytes-like object, interpreting its content as machine values. |
| 2      | `tobytes()`          | Converts the array to a bytes representation.                              |
| 3      | `fromfile(f, n)`     | Reads `n` items from file object `f` and appends them to the array.        |
| 4      | `tofile(f)`          | Writes all items to the file object `f`.                                   |
| 5      | `fromlist(list)`     | Appends items from the list to the array.                                  |
| 6      | `tolist()`           | Converts the array to a list with the same items.                          |
| 7      | `fromunicode(s)`     | Extends the array with data from Unicode string `s`. Requires type code `'u'`. |
| 8      | `tounicode()`        | Converts the array to a Unicode string. Requires type code `'u'`.         |
